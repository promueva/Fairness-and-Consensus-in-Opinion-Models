```python
# third one, the big one

import numpy as np
from PushDeGrootGraph import *
from WordGenerator import *
import matplotlib.pyplot as plt

graph_samples = 40
trials_per_graph = 1
nodes = 5000
word_len = 100000
distances = []

for s in range(graph_samples):
    print("system "+str(s))
    skip = False
    distance = []

    G = PushDeGrootGraph(nodes, keep_history=False)
    G.set_initial_opinions()
    w = generate_random_word(G, word_len)
    for t, e in enumerate(w):
        if not skip:
            G.make_call(G.labels[e])
            d=max(G.opinion.values())-min(G.opinion.values())
            if d < 0.01:
                skip = True
        distance.append(d)
    
    distances.append([*distance])

avg = []
for (t, ds) in enumerate(np.transpose(distances)):
    avg.append(np.average(ds))

# Create a figure and axis
fig, ax = plt.subplots()

ax.plot(avg)


plt.show()
```

```python


with open("completedata_randomRuns3.txt", "w") as output:
  output.write("t\td\n")

  for (t, data) in enumerate(avg):
    #if t % 100 == 0:
      output.write(str(t)+"\t"+str(data)+"\n")

print(data)
```

```python
# Edge fairness counter example: Cycle of 3 agents
import numpy as np
from PushDeGrootGraph import *
from WordGenerator import *
import matplotlib.pyplot as plt

from PushDeGrootGraph import *

G = PushDeGrootGraph(3, ops=[0,1])
G.draw()

w = "abcabbcabbbc"
w = ""
i=1
while len(w) < 500:
    w = w + i*"a"
    w = w + (i+1)*"b"
    w = w + (i+2)*"c"
    i = i + 1

print(w)

G.execute_word(w)


history = np.transpose(G.history)

# Create a figure and axis
fig, ax = plt.subplots()

# Iterate through each column (representing 'i') in the transposed 'history' and plot it
for i, data in enumerate(history):
    ax.plot(data,  label=f"i = {i+1}")


#plt.show()


```
