```python
# third one, the big one

import numpy as np
from PushDeGrootGraph import *
from WordGenerator import *
import matplotlib.pyplot as plt

graph_samples = 40
trials_per_graph = 1
nodes = 5000
word_len = 100000
distances = []

for s in range(graph_samples):
    print("system "+str(s))
    skip = False
    distance = []

    G = PushDeGrootGraph(nodes, keep_history=False)
    G.set_initial_opinions()
    w = generate_random_word(G, word_len)
    for t, e in enumerate(w):
        if not skip:
            G.make_call(G.labels[e])
            d=max(G.opinion.values())-min(G.opinion.values())
            if d < 0.01:
                skip = True
        distance.append(d)
    
    distances.append([*distance])

avg = []
for (t, ds) in enumerate(np.transpose(distances)):
    avg.append(np.average(ds))

# Create a figure and axis
fig, ax = plt.subplots()

ax.plot(avg)


plt.show()
```

```python
import numpy as np
from PushDeGrootGraph import *
from WordGenerator import *
import matplotlib.pyplot as plt

# Edge fairness counterex: 
# Validation: The subsequences of lows and highs of agent 2 are monotonically decreasing and increasing


from PushDeGrootGraph import *

g = nx.DiGraph([(0,1),(1,0),(2,1),(1,2)])

G = PushDeGrootGraph(4, ops=[0,0.5,1], G=g, keep_history=False)

total_rounds = 100000

prev_min = 0.5
prev_max = 0.5
i=1

while i < total_rounds:
    G.execute_word("a"*i)
    if prev_min < G.opinion[1]:
        print(f"min does not decrease monotonically in round {i}")

    G.execute_word("b")

    G.execute_word("d"*i)
    if prev_max > G.opinion[1]:
        print(f"max does not increase monotonically in round {i}")

    G.execute_word("c")

    i = i + 1
    if i % (total_rounds / 20) == 0:
        print(f"checked {i} rounds")
```
